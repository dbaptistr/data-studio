How many crimes do mexican citizens commit outside of Mexico? 
The Mexican Ministry of Foreign Affairs has a public database showing the amount of criminal cases for which it has provided protection and consular assitance in the last seven years.
The datasets are for crimes inside and outside the United States.

(https://datos.gob.mx/busca/dataset/penal--casos-de-proteccion-y-o-asistencia-consular-atendidos-por-la-rdcm-en-el-mundo)

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

```python
df = pd.read_csv('CasosdeProteccionyAsistenciaConsularRegistradosporlaRDCMx-Penal.csv')

df.drop(['�MBITO'], axis=1, inplace=True)
df
```

```python
df_non = pd.read_csv('Casos_noEU.csv')
df_non
```

```python
df = df.rename(columns={'RUBRO DE PROTECCI�N': 'crime'})
df.dtypes
```

```python
df = df.fillna(0)
df
```

```python
df['Total_2010'] = df['DESCONOCIDO_2010'] + df['HOMBRE_2010'] + df['MUJER_2010']
df
```

```python
df['Total_2011'] = df['DESCONOCIDO_2011'] + df['HOMBRE_2011'] + df['MUJER_2011']
df['Total_2012'] = df['DESCONOCIDO_2012'] + df['HOMBRE_2012'] + df['MUJER_2012']
df['Total_2013'] = df['DESCONOCIDO_2013'] + df['HOMBRE_2013'] + df['MUJER_2013']
df['Total_2014'] = df['DESCONOCIDO_2014'] + df['HOMBRE_2014'] + df['MUJER_2014']
df['Total_2015'] = df['DESCONOCIDO_2015'] + df['HOMBRE_2015'] + df['MUJER_2015']
df['Total_2016'] = df['DESCONOCIDO_2016'] + df['HOMBRE_2016'] + df['MUJER_2016']
df['Total_2017'] = df['DESCONOCIDO_2017'] + df['HOMBRE_2017'] + df['MUJER_2017']
```

```python
df
```

```python
df.sort_values(by='Total_2010', ascending=False)
```

```python
new_df = df[['crime','Total_2010', 'Total_2011', 'Total_2012', 'Total_2013', 'Total_2014', 'Total_2015', 'Total_2016', 'Total_2017']].copy()
```

```python
new_df = new_df.sort_values(by='Total_2010', ascending=False)
new_df
```

```python
new_df.plot(x='crime', y='Total_2010', kind='barh')
new_df.plot(x='crime', y='Total_2011', kind='barh')
new_df.plot(x='crime', y='Total_2012', kind='barh')
new_df.plot(x='crime', y='Total_2013', kind='barh')
new_df.plot(x='crime', y='Total_2014', kind='barh')
new_df.plot(x='crime', y='Total_2015', kind='barh')
new_df.plot(x='crime', y='Total_2016', kind='barh')
new_df.plot(x='crime', y='Total_2017', kind='barh')
```
