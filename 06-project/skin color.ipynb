```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

# Skin color ID

```python
skinc= pd.read_csv('conjunto_de_datos_tcoe_enadis_2017.csv')
```

```python
skinc.head()
```

```python
skinc2 = skinc['po5_1'].value_counts(dropna=False)
```

```python
skinc2 = pd.DataFrame(data=skinc2)
```

```python
skinc2
```

```python
skinc2['po5_1'].sum()
```

```python
skinc2['Percent'] = (skinc2['po5_1'] * 100) / skinc2['po5_1'].sum()
```

```python
skinc2
```

```python
skinc2.plot(y='Percent',kind='barh')
```
